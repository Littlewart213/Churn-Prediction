**Artifical Neural Network (ANN)**

```python
X_train_tensor = torch.tensor(X_Train_final.values, dtype=torch.float32)
y_train_tensor = torch.tensor(Y_Train.values, dtype=torch.float32).unsqueeze(1)
```

```python
X_test_tensor = torch.tensor(X_Test_final.values, dtype=torch.float32)
```

```python
class ANN_Model(nn.Module):
    def __init__(self, input_dim):
        super(ANN_Model, self).__init__()
        self.fc1 = nn.Linear(input_dim, 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, 1)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        return x
```

```python
input_dim = X_train_tensor.shape[1]
model = ANN_Model(input_dim)
```

```python
criterion = nn.BCELoss()  # Binary cross-entropy loss for binary classification
optimizer = optim.Adam(model.parameters(), lr=0.000001)  # Adam optimizer
```

```python
epochs = 30
batch_size = 16
```

```python
def get_batches(X, y, batch_size):
    for i in range(0, len(X), batch_size):
        X_batch = X[i:i+batch_size]
        y_batch = y[i:i+batch_size]
        yield X_batch, y_batch
```

```python
for epoch in range(epochs):
    model.train()
    epoch_loss = 0
    for X_batch, y_batch in get_batches(X_train_tensor, y_train_tensor, batch_size):
        optimizer.zero_grad()
        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f'Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}')
```

```python
model.eval()
with torch.no_grad():
    y_test_pred = model(X_test_tensor)
    predicted_probabilities_ANN = y_test_pred.numpy()
```

```python
predictions_df_ANN = pd.DataFrame({
    'CustomerID': test_df['CustomerID'],
    'predicted_probability': predicted_probabilities_ANN.flatten()
})
predictions_df_ANN
```
